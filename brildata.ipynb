## *Common Data Format*

The specification of the data exchange format on bril b2in eventing infrastructure is defined in:

**daq/interface/bril/include/interface/bril/DataFormat.h**

The following concepts are implemented in the header file.

````python
%cat $workdir/daq/interface/bril/include/interface/bril/DataFormat.h
````

### *data format version agreement*

The sender and receiver of the bril eventing data must agree on the version of the definition. 

The sender must send the string defined by bril::DATA_VERSION as property **"DATA_VERSION"** metadata (property list) 
of the data package. 

The receiver must check the property list of the arriving data. If "DATA_VERSION" is absent or its value does not match
the one defined by the header file, the package should not be considered as bril data and discarded.

**"DATA_VERSION"** is the only property metadata required and recognized by the brildaq components. 

### *topics*

A topic is the communication tag for a collection of data packages of the same type and size. They are sent from the same bril software component, which is not necessarily the same application or the same readout unit. The packages exchanged with the same topic name can from different algorithm, group type, readout or application instance. Such variations are tagged in the specific bits of the data package header.
 
In the header file the following seven data sources are defined whose id is defined as enum: **bril::DataSource**

  **TCDS=1, DIP, PLT, BCM1F, HF, BHM, LUMI**
  

  
### *data header*

Data packages exchanged on the bril eventing must carry the same header where the time, provenance, frequency and data size information are provided.

```
 struct DatumHead{
    unsigned int runnum;
    unsigned int lsnum;
    unsigned int cmslsnum;
    unsigned int nbnum;
    unsigned short publishnnb;
    unsigned short accumulatennb;
    unsigned int resourceidvalue;
    unsigned int timestampsec;
    unsigned int timestampmsec;
    unsigned int totalsize;
    unsigned char payloadanchor[1];   
    }
```

DatumHead struct is at the beginning address of the package followed by the payload starting from address payloadanchor.

**runnum, lsnum, nbnum** are acquired time fields. Two other time fields are *timestampsec, timestampmsec*. 

**timestampsec** is the unix timestamp (GMT) counting seconds since epoch. 

**timestampmsec** counts the additional milli-seconds. 

If the timestamps are not provided by the hardware, you can fill them by yourself in the source.

**cmslsnum** has meaning only for bril component related or listen to cms cdaq. Please set to 0 if not connected to cms.  

**publishnnb** is the data publication frequency in term of lumi nibble.

**accumulatennb** is a hint that can be set by the data source to the data processor on how many nibbles should be considered full statistics. Please set to 0 if not applicable.

**totalsize** is the total size (header+payload) of the data package. 

**resourceidvalue** contains the compressed information about the provenance of the data package. The 32 bits are partitioned as follows:

|      | DataSourceId  | AlgoId  | VarGroupId | ReadoutId | InstanceId | StorageTypeID |
|:----:|:-------------:|:-------:|:----------:|:---------:|:----------:|:-------------:|
|n bits| 8 bits    | 4 bits  | 4 bits     |  8 bits   | 4 bits     | 4 bits        |
|value range | [1,255] | [1,15]  | [1,15]     | [1,255]   | [1,15]     | [1,15]        |      

0 is a special reserved value for invalid in all above fields.

*DataSourceId* is the sending bril component as explained above.

*AlgoId* distinguishes the same type of data coming from or for different algorithms, such as beam1 or beam2; occupancy or energy sum 

*VarGroupId* is the second dimension of freedom of the same type of data package, such as value or error. 

*ReadoutId* is the readout unit id of the same-type data, such as channel number

*InstanceId.* Sometimes when reading out, one can use the application instance id as another level of grouping in addition to the readout id. (It's however not clear currently if this field is necessary here).

*StorageTypeID* In the bril data header, the following storage types are defined and acknowledged whose id is defined as enum: bril::StorageType

**UINT8=1, INT8, UINT16, INT16, UINT32, INT32, UINT64, INT64, FLOAT32, DOUBLE64, STRING**

For boolean type, one can just use UINT8. 


DatumHead class provides setter and getter functions for above fields:

  ```
    setTime
    setResource
    setFrequency
    setTotalsize
    getDataSourceID
    getInstanceIDL
    getStorageTypeID
    getReadoutIDString
    getInstanceIDString
    totsize
  ```

### *data payload*

Data package payload is an array of the above allowed premitive types. A single value is considered an array of one element.

DatumHead class provides the following functions about payload.
   ```
    payloadsize
    nitems
   ```
   
   nitems is the number of items allocated in the array.
   

### *define a topic and the associated data*

macro DEFINE_DATATOPIC defines a topic and its data at the same time.
One example can be found in BHMTopics.h:

   ```
 DEFINE_DATATOPIC(bhmocchist,unsigned int,BHM_BINPERBX*MAX_NBX,NULL,NULL,BHM_NRU,"bhm occupancy histogram","");
   ```

This line defines a topic "bhmocchist" whose payload is **unsigned int[BHM_BINPERBX*MAX_NBX]**. 

The 4th, 5th arguments are the address of the lookup tables for algorithm and vargroup variations. Since not applicable for this topic, they are set to NULL. The 6th argument is the number of readouts. The last two arguments are title and unit string.


   ```
 DEFINE_DATATOPIC(bhmbkg,float,1,BHMBkgAlgos::lookuptable,NULL,NULL,"bhm background rate","Hz/10^11 protons");
   ```
   
This line defines a topic "bhmbkg" whose payload is a single float. It can be a bhmbkg number for algorithm BHMBkgAlgos::BEAM1 or 
BHMBkgAlgos::BEAM2. The algorithm enum and its lookup table is defined by :
   ```
      #define BHMBkgAlgos_ENUM(XX) XX(BEAM1,=1) XX(BEAM2,)
      DEFINE_LOOKUPTABLE(BHMBkgAlgos,BHMBkgAlgos_ENUM)
   ```
   

### *acquire static information about a topic at run time*

At run time, **static** helper functions bound to namespace "bril::Datum &lt;typename T&gt;", whose alias is "bril::topicnameT", can be used to to acquire the size and descrption information about a topic and its data because the calculations are done at compile time. For example for topic "bhmbkg", the following static helper functions are available:
   ```
      bril::bhmbkgT::topicname()
      bril::bhmbkgT::storagetype()
      bril::bhmbkgT::n()
      bril::bhmbkgT::maxsize()
      bril::bhmbkgT::description()
      bril::bhmbkgT::unit()
      bril::bhmbkgT::algo_lookuptable()
      bril::bhmbkgT::group_lookuptable()
      bril::bhmbkgT::nru()
   ```
   
   maxsize() is particularly useful when you need to allocate the correct size of a memory buffer.
   

### *manipulate received message buffer*


When received a message buffer, we know the topic name, e.g. bril::bhmbkgT::topicname(), and the buffer address void\*. In order to access the buffer content, we need to cast the void\* to the desired shape. If only interested in the general information, we just cast it into DatumHead\*
  ```
      bril::DatumHead* head = (bril::DatumHead*)buffer;
      unsigned int runnum = head->runnum;
  ```

In order to access the payload portion of the buffer, we need to cast the buffer into topicnameT* first.
   ```
      bril::bhmbkgT* full = (bril::bhmbkgT*)buffer;
      float bkgnum = full->payload()[0];
   ```
  
You can access the payload element by index because operator [] of the payload object is overloaded.  

## *MySource and MyProcessor*

In this example, typical usage of data macros and helper functions are shown.

Note: you only need to include yourTopics.h. Including other headers introduces unnecessary dependency.

MySource publishes BHM occupancy and amplitude histograms defined in BHMTopics.h on nibble frequency to eventing bus "brildata". 

MyProcessor receives and accumulates histograms from eventing bus "brildata" for 64 nibbles. When data from all 40 channels and 64 nibbles are collected, the processor calculates the beam background results and publishes them to the eventing.

Two boundary conditions in time are handled:

    If the first received lumi section does not have the full statistics, discard.

    If the last received lumi section does not have the full statistics, discard.

### *write MyProcessor*

````python
# MyProcessor.h
# The processor receives and accumulates histograms from bhmocchist, bhmamphist on nibble frequency for every channel.
# It checks the statistics on the channels and nibbles. When statistics are full, process the collection to get beam background
# numbers and publish them
# We put the collected data temporarily in the cache m_occhistcache, m_amphistcache, indexed by channel id.
# The channel content item is an internal struct InData which keeps a nibble counter and the memory reference to the 
# base accumulating histogram.
#
# Methods check_channelstatistics, check_nbstatistics checks if the accumulated statistics in terms of number of channels 
# and number of nibbles are full.
# 
# Methods occ_accumulateInChannel, occ_accumulateInChannel accumulate received histogram of the specified channel to its base 
# histogram in cache.
#
# When the statistics are judged to be full, do_publish and clear_cache are called
#
# We want to check the boundary condition where the last nibble received is not the 64th of the lumi section, 
# i.e. the last lumi section is not full. We need to check regularly the timestamp of the last received message. 
# Hense we inherit from toolbox::task::TimeListener and implement method timeExpired.
%cat $workdir/daq/bril/dummysource/include/bril/dummysource/MyProcessor.h
````

````python
# MyProcessor.cc
# In eventing::api::BusReadyToPublish callback, we subscribe to the bhm histogram topics 
#          this->getEventingBus("brildata").subscribe(bril::bhmocchistT::topicname())
#          this->getEventingBus("brildata").subscribe(bril::bhmamphistT::topicname())
# Note the usage of the helper functions instead of hardcoded strings here.
# Here we also set up and launch the timer for checking the age of the last received message
# 
# In the b2in message receiving callback onMessage method, we check the "DATA_VERSION" property of the received message
# If it's empty or mismatch what we have included in the header file, we ignore the message.
#
#   Then we check the status of the accumulated data in the cache (Note: not including the current message), 
# if statistics full, process and publish the result for the data in cache, clear the cache.
# In the case if on the lumi section boundary, previous collected data do not have full statistics, we discard and clear the cache.
# 
#   Then we look at the currently received message, adding the content into its corresponding base histogram 
# or make a new base histogram entry in the cache.
# 
# In do_process method, we do some bx, channel aggregation and averaging work to obtain two final beam background numbers.
# We fill the outgoing package headers and payload.
# Note: how to fill and manipulate the payload content in the memory pool buffer directly without memcpy.
#
# In do_publish method, we publish the result package to the "bhmbkg" topic as different algorithm variations
# bril::BHMBkgAlgos::BEAM1, bril::BHMBkgAlgos::BEAM2
#
# In the timer callback method timeExpired, if the cached data is stale we process and publish the result in case of full statistics
# otherwise, discard and clear cache.
%cat $workdir/daq/bril/dummysource/src/common/MyProcessor.cc
````
